```python
# Sinc
def sinc(x, w):
    return np.sin(w*x) * 1/x

def sinc2d(x, wx, y, wy):
    return sinc(x, wx) * sinc(y, wy)
```

```python
# 2D Gaussian
def ABCGauss(x, y, A, x0, y0, a, b, c):
    return A * np.exp(a * (x - x0) ** 2 + b * (x - x0)*(y - y0) + c * (y - y0) ** 2)

def SigmaThetaGauss(x, y, A, x0, y0, sx, sy, theta):
    a = (np.cos(theta) ** 2 / (2 * sx ** 2)) + (np.sin(theta) ** 2 / (2 * sy ** 2))
    b = -(np.sin(theta) * np.cos(theta))/(2 * sx ** 2) + (np.sin(theta) * np.cos(theta))/(2 * sy ** 2)
    c = (np.sin(theta) ** 2 / (2 * sx ** 2)) + (np.cos(theta) ** 2 / (2 * sy ** 2))

    return ABCGauss(x, y, A, x0, y0, a, b, c)
```

```python
fig1, ax1 = plt.subplots(figsize=(7, 7))

xlin = np.linspace(-10, 10, 100)
ylin = np.linspace(-10, 10, 100)

xx, yy = np.meshgrid(xlin, ylin)

# x, y, A, x0, y0, sx, sy, theta
zz = ABCGauss(xx, yy, 10, 0, 0, 1, 0, 0)

ax1.pcolormesh(xx, yy, zz)
ax1.set_aspect('equal')
```
