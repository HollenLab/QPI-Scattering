```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating rectangular lattice
z = np.zeros((100, 100))

for i in range(0, 100, 5):
    for j in range(0, 100, 5):
        z[i][j] = 1
```

```python
fig, ax = plt.subplots()

ax.pcolormesh(z)
ax.set_aspect('equal')
ax.scatter(40, 40)
ax.scatter(40, 45) #spacing is 5 units
```

```python
ftz = np.fft.fft2(z)
ftz = np.fft.fftshift(ftz)
```

```python
fig2, ax2 = plt.subplots()

stepl = 1
fx = np.fft.fftfreq(100, stepl)
fy = np.fft.fftfreq(100, stepl)
fxs = np.fft.fftshift(fx)
fys = np.fft.fftshift(fy)

ax2.pcolormesh(fxs, fys, np.abs(ftz))
ax2.scatter(0, 0)
ax2.scatter(1/5, 0)
```
