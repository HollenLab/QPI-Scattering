```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def hann1d(x, N, x0):
    val = 0

    #if (np.abs(x-x0) < N/2):
    #    val = np.cos(np.pi * (x-x0)/N) ** 2

    val = np.cos(np.pi * (x-x0)/N) ** 2
    return val
```

```python
fig, ax = plt.subplots()

N = 1000
xlin = np.linspace(-10, 10, N)

y = np.array([hann1d(x, 30, -1) for x in xlin])

ax.plot(xlin, y)
```

```python
def hann2d(x, y, Nx, Ny, x0, y0): 
    val = hann1d(x, Nx, x0) * hann1d(y, Ny, y0)

    return np.where((np.abs(x-x0) <= 0.5 * Nx) & (np.abs(y-y0) <= 0.5 * Ny), val, 0)

hann_vec = np.vectorize(hann2d)
```

```python
fig1, ax1 = plt.subplots()

N = 1000
xlin = np.linspace(-5, 5, N)
ylin = np.linspace(-5, 5, N)

xx, yy = np.meshgrid(xlin, ylin)
z = hann2d(xx, yy, 1,5, 0,0)
#z = mask(xx, yy)

pc = ax1.pcolormesh(xlin, ylin, z)
fig1.colorbar(pc, ax=ax1)
ax1.set_aspect('equal')
```
