```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
signal = np.array([-2, 8, 6, 4, 1, 0, 3, 5], dtype=float)
fourier = np.fft.fft(signal)
n = signal.size

timestep = 0.1
timelin = np.arange(0, 0.1*n, step=0.1)
freq = np.fft.fftfreq(n, d=timestep)
freq = np.fft.fftshift(freq)

timeagain = np.fft.ifft(fourier)
timelinagain = np.fft.fftfreq(n, d=(freq[1]-freq[0]))
print(timelinagain)
print(timelin)
```

```python
print("n", n)
print("dt:", timestep)
print("dF:", freq[1]-freq[0])
```

```python
dF = freq[1]-freq[0]
dt_calc = 1/(dF * n)
print(dt_calc)
tlin_calc = np.arange(0, dt_calc * n, step=dt_calc)
print(tlin_calc)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(12, 4))

axs[0].plot(timelin, signal)
axs[0].set_xlabel("Time (s)")

axs[1].plot(freq, np.abs(fourier), marker='.', linestyle='none')
axs[1].set_xlabel('Angular Frequency (s^-1)')

axs[2].plot(tlin_calc, np.abs(timeagain), marker='.', linestyle='solid')
axs[2].set_xlabel('Time (s)')
```
