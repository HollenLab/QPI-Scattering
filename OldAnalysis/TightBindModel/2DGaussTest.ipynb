```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, y, x0, y0):
    return np.exp(-((x-x0) ** 2 + (y-y0) ** 2))
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

xx, yy = np.meshgrid(x, y)

zz = f(xx, yy, 2, 0)
```

```python
fig, ax = plt.subplots()

ax.pcolormesh(xx, yy, zz)
ax.set_aspect('equal')
```
